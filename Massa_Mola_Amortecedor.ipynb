```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def sistema_mma(y, t, m, b, k, f):
    x1, x2 = y
    u = np.sin(2 * np.pi * f * t)
    dx1dt = x2
    dx2dt = (-k * x1 - b * x2 + u) / m
    return [dx1dt, dx2dt]

m = np.random.uniform(0.1, 10.0)
b = np.random.uniform(0.1, 1.0)
k = np.random.uniform(1.0, 10.0)
x0 = np.random.uniform(-1.0, 1.0, size=2)
f = 5.0
t = np.linspace(0, 100, 1000)

solucao = odeint(sistema_mma, x0, t, args=(m, b, k, f))

x1 = solucao[:, 0]
x2 = solucao[:, 1]

plt.plot(t, x1, label='x1(t)')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(t, x2, label='x2(t)')
plt.xlabel('tempo')
plt.ylabel('x2')
plt.legend()
plt.grid()
plt.show()
```
