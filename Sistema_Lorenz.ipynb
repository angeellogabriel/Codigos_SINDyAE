```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def sistema_lorenz(z0, t, sigma=10, beta=8/3, rho=28):
  f = lambda z, t: [sigma*(z[1] - z[0]),
                    z[0]*(rho - z[2]) - z[1],
                    z[0]*z[1] - beta*z[2]]

  z = odeint(f, z0, t)
  z1 = z[:, 0]
  z2 = z[:, 1]
  z3 = z[:, 2]

  plt.plot(z1, z2, label='z1')
  plt.xlabel('tempo')
  plt.ylabel('v√°riaveis')
  plt.legend()
  plt.grid()
  plt.show()

z0 = [1.0, 0.0, 0.0]
t = np.linspace(0, 50, 10000)
sistema_lorenz(z0, t)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def lorenz(z0, t, sigma=10, beta=8/3, rho=28):
    f = lambda z, t: [sigma*(z[1] - z[0]),
                      z[0]*(rho - z[2]) - z[1],
                      z[0]*z[1] - beta*z[2]]

    z = odeint(f, z0, t)

    z1 = z[:, 0]
    z2 = z[:, 1]
    z3 = z[:, 2]

    return z1, z2, z3

z0 = [1.0, 0.0, 0.0]
t = np.linspace(0, 50, 10000)

z1, z2, z3 = lorenz(z0, t)

plt.plot(z1, z2)
plt.xlabel('z1')
plt.ylabel('z2')
plt.title('z1 vs. z2')
plt.show()
```

```python
plt.plot(z1, z3)
plt.xlabel('z1')
plt.ylabel('z3')
plt.title('z1 vs. z3')
plt.show()
```

```python
plt.plot(z2, z3)
plt.xlabel('z2')
plt.ylabel('z3')
plt.title('z2 vs. z3')
plt.show()
```

```python
#3D
import numpy as np
import matplotlib.pyplot as plt

def lorenz(xyz, sigma=10, beta=8/3, rho=28):
  x, y, z = xyz
  return np.array([sigma * (y - x),
                   rho * x - y - x * z,
                   x * y - beta * z])

passos = 10000
xyz = np.zeros((passos, 3))
xyz[0] = [0, 1, 1.05]
dt = 0.01

for i in range(1, passos):
  xyz[i] = xyz[i-1] + lorenz(xyz[i-1])*dt

plt.figure(figsize=(8,8)).add_subplot(projection='3d').plot(*xyz.T)
plt.show()
```
